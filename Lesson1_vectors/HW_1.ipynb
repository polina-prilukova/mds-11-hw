# Домашнее задание "Линейная алгебра. Векторы".

Даны вектора x = (1,1), y = (2, 0) и z = (0,2)

Надо:
- Изобразить вектора на экране
- Изобразить точку x + y + z
- Найти угол между векторами x и y



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 1])
y = np.array([2, 0])
z = np.array([0, 2])
```

```python
# Пункт 1. Изобразить вектора на экране
plt.figure(figsize=(7, 7))
plt.plot(x[0], x[1], 'o')
plt.plot(y[0], y[1], 'o')
plt.plot(z[0], z[1], 'o')
plt.grid(True)
```

```python
# или так, в более привычном геометрическом представлении
zero_x = [0, 0, 0]
zero_y = [0, 0, 0]
x_coords = [x[0] for x in np.concatenate([x, y, z]).reshape(3, 2)]
y_coords = [x[1] for x in np.concatenate([x, y, z]).reshape(3, 2)]
plt.quiver(zero_x, zero_y, x_coords, y_coords, color=['r','g','b'], angles='xy',  scale_units='xy', scale=1)
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.grid(True)
plt.show()
```

```python
# Пункт 2. Изобразить точку x + y + z
# её координатами будет поэлементная сумма координат наших векторов
# точку - сумму векторов обозначаю красным

sum_coords = x + y + z
plt.figure(figsize=(7, 7))
plt.plot(sum_coords[0], sum_coords[1], 'o', markersize=10, color='r')
plt.plot(x[0], x[1], 'o')
plt.plot(y[0], y[1], 'o')
plt.plot(z[0], z[1], 'o')
plt.grid(True)
```

```python
# или так
zero_xx = [zero_x[0], zero_x[0]+x[0], zero_x[0]+x[0]+y[0]]
zero_yy = [zero_x[1], zero_x[1]+x[1], zero_x[1]+x[1]+y[1]]
plt.quiver(zero_xx, zero_yy, x_coords, y_coords, angles='xy',  scale_units='xy', scale=1)
plt.quiver(0, 0, sum_coords[0], sum_coords[1], color='r', angles='xy',  scale_units='xy', scale=1)
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.grid(True)
plt.show()
```

```python
#  Пункт 3. Найти угол между векторами x и y

# найдем косинус угла между векторами как частное от деления скалярного произведения векторов на проиведение их длин

x_length = np.linalg.norm(x)
y_length = np.linalg.norm(y)
cosine = np.dot(x, y)/(x_length * y_length)
angle = np.degrees(np.arccos(cosine))
print(f'Угол между векторами х и у равен {round(angle, 2)} градусам')
```
