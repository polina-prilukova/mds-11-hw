### Уровень 0:


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.



```python
import numpy as np
import matplotlib.pyplot as plt
```

Первая производная будет равна:  
$f'(x) = 5x^4 + 4cos(2x)*2 + (-sin(3x+3)*3) = 5x^4 + 8cos(2x) - 3sin(3x+3)$ 

Вторая производная будет равна:  
$f''(x) = 20x^3 + 8*(-sin(2x)*2) - 3*(cos(3x+3)*3) = 20x^3 - 16sin(2x) - 9cos(3x+3)$

```python
# в точке x = 1 первая и вторая производная примут следующие значения
x = 1
d1 = 5*(x**4) + 8*np.cos(2*x) - 3*np.sin(3*x + 3)
d2 = 20*(x**3) - 16*np.sin(2*x) - 9*np.cos(3*x + 3)

print(f"f'(1) = {round(d1, 4)}\nf''(1) = {round(d2, 4)}")
```
